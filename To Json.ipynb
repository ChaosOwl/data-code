```python
import pandas as pd
import json
```

```python
dataframe = pd.read_excel('suggested job for seniors.xlsx')
```

```python
for i in range(len(dataframe)):
    list_of_dict = []
    if type(dataframe['Skill'][i]) == str:
        list_of_skill = dataframe['Skill'][i].split(',')
        for item in list_of_skill:
            dict_of_skill = {'name':item}
            list_of_dict.append(dict_of_skill)
    dataframe['Skill'][i] = list_of_dict
            
for i in range(len(dataframe)):    
    list_of_dict = []
    if type(dataframe['Certification'][i]) == str:
        list_of_Certification = dataframe['Certification'][i].split(',')
        for item in list_of_Certification:
            dict_of_Certification = {'name':item}
            list_of_dict.append(dict_of_Certification)
    dataframe['Certification'][i] = list_of_dict
```

```python
jsonlist = []
for i in range(len(dataframe)):
    aItem = {}
    aItem["id"] = i
    aItem["job_name"] = dataframe['Job'][i]
    aItem["category"] = dataframe['Category'][i]
    aItem['skills'] = dataframe['Skill'][i]
    aItem['certification'] = dataframe['Certification'][i]
    aItem['desc'] = dataframe['Description'][i]
    aJson = json.dumps(aItem)
    jsonlist.append(aItem)
```

```python
savefile = 'joblist-large.json'

with open(savefile, 'w') as f:
    f.write(json.dumps(jsonlist))
```
